```python
SalesData = pd.read_csv("./Atelier 1/salesData.csv")
```

```python
SalesData.head()
```

```python
profit = SalesData["total_profit"]
```

```python
#QUESTION 1
ax = plt.axes()
ax.set_ylim([100000,500000])
ax.set_xticks(range(1,13))
ax.set_yticks(range(100000,500001,100000))
ax.set_xlabel('Month Number')
ax.set_ylabel('Profit in Dollars')
plt.title('Company Profit Per Month')
plt.plot(SalesData['month_number'], profit)
plt.savefig('Figures/CR_1.jpg')
plt.show()
```

```python
#QUESTION 2
ax = plt.axes()
ax.set_xlabel('month_number')
ax.set_ylabel('Profit in dollar')
plt.title('Company Sales data of last year')
plt.plot(range(1,13), profit,linewidth=3, color='r', linestyle=':', marker='o', markerfacecolor='k')
plt.legend(['Company Sales data of last year'], loc=4)
plt.savefig('Figures/CR_2.jpg')
```

```python
#QUESTION 3
ax = plt.axes()
ax.set_xlabel('Month Number')
ax.set_ylabel('Sales Units in number')
ax.set_xticks(range(1,13))
ax.set_ylim([1000,18000])
for column in SalesData.columns[1:-2]:
    plt.plot(range(1,13), SalesData[column], marker='o', linewidth=2)
plt.legend(SalesData.columns[1:-2]+' Sales Data', loc=2)
plt.title('Sales Data')
plt.savefig('Figures/CR_3.jpg')
```

```python
ax = plt.axes()
ax.set_xlabel('Month Number')
ax.set_ylabel('Number of units sold')
ax.set_xticks(range(1,13))
plt.scatter(range(1,13),SalesData["toothpaste"])
plt.legend(['toothpaste Sales Data'], loc=2)
plt.grid()
plt.title('Toothpaste Sales Data')
plt.savefig('Figures/CR_4.jpg')
```

```python
#QUESTION 5
profits = []
for column in SalesData.columns[1:-2]:
    profit = sum(SalesData[column])
    profits.append(profit)

plt.pie(profits, labels=SalesData.columns[1:-2], autopct='%1.1f%%')
plt.legend(SalesData.columns[1:-2], loc=4)
plt.title('Sales Data')
plt.savefig('Figures/CR_5.jpg')
```
